```python
# import libraries (including our short d3_lib script)
from IPython.core.display import HTML
import d3_lib, random
import re
import json
```

```python
%%script bash
arq --results JSON --data https://raw.githubusercontent.com/Planteome/plant-trait-ontology/master/plant-trait-ontology.obo.owl '
PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>  
PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>  
PREFIX owl: <http://www.w3.org/2002/07/owl#> 
PREFIX oio: <http://www.geneontology.org/formats/oboInOwl#>  
PREFIX obo: <http://purl.obolibrary.org/obo/>  

#CONSTRUCT 
#{
#    ?x rdfs:subClassOf ?o .
#}
SELECT ?x ?xlabel ?o ?olabel
WHERE
{
    ?x rdfs:subClassOf ?o .
    filter(!isblank(?o)) 
    filter(!regex(str(?o), "PATO"))
    filter(regex(str(?o), "TO"))
    ?x rdfs:label ?xlabel . 
    ?o rdfs:label ?olabel .
}' > to.json
```

```python
with open('to.json') as data_file:    
    data = json.load(data_file)
    
    list = data["results"]["bindings"]
    ####create nodes
    nodes=[]
    for entry in list:
        nodes.append( {"name":entry["xlabel"]["value"], "parent": entry["olabel"]["value"] })
        
        
    #add the root
    nodes.append( {"name":"plant trait" })
```

```python
with open('graphDnd.json', 'w') as outfile:
    json.dump(nodes, outfile)
```

```python
# visualize as force-directed graph in D3
HTML(d3_lib.set_styles(['dndTree']) + 
'<script src="lib/d3/d3.min.js"></script>' + 
'<script src="http://code.jquery.com/jquery-1.10.2.min.js"></script>' + 
'<script src="js/dndTree.js"></script>' + 
      d3_lib.draw_graph('dndTree', {'data': nodes}) )
```
